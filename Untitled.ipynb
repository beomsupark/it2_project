```python
from PIL import Image

def image_crop( infilename , save_path):
    """
    image file 와 crop한이미지를 저장할 path 을 입력받아 crop_img를 저장한다.
    :param infilename:
        crop할 대상 image file 입력으로 넣는다.
    :param save_path:
        crop_image file의 저장 경로를 넣는다.
    :return:
    """
 
    a_img = Image.open( infilename )
    img= a_img.convert('RGB')
    (img_h, img_w) = img.size
    print(img.size)
 
    # crop 할 사이즈 : grid_w, grid_h
    grid_w = img_w/4 # crop width
    grid_h = img_h/4 # crop height
    print("grid",grid_w)
    range_w = (int)(img_w/grid_w)
    range_h = (int)(img_h/grid_h)
    #print(range_w, range_h)
 
    i = 0
 
    for w in range(range_w):
        for h in range(range_h):
            bbox = (h*grid_h, w*grid_w, (h+1)*(grid_h), (w+1)*(grid_w))
            #print(h*grid_h, w*grid_w, (h+1)*(grid_h), (w+1)*(grid_w))
            # 가로 세로 시작, 가로 세로 끝
            crop_img = img.crop(bbox)
 
            fname = "{}.jpg".format("{0:05d}".format(i))
            savename = save_path +fname
            crop_img.save(savename)
            #print('save file ' + savename + '....')
            i += 1

            
```

```python

import numpy as np
import matplotlib.pylab as plt

def compare_image(a_filename , b_filename):
    imgs = []
    imgs.append(cv2.imread(a_filename ))
    imgs.append(cv2.imread(b_filename))

    hists = []

    for img in imgs:
        hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
        hist = cv2.calcHist([hsv], [0, 1], None, [180, 256], [0, 180, 0, 256])
        cv2.normalize(hist, hist, 0, 1, cv2.NORM_MINMAX) 
        hists.append(hist)
        
    query = hists[0]
    ret = cv2.compareHist(query, hists[1], cv2.HISTCMP_BHATTACHARYYA)
    
    if(ret>0.9):
        return True
    return False
```

```python
import numpy as np
from PIL import Image

def concat(counted, save_path):
    paths = [os.path.join('transform_image', x) for x in os.listdir('transform_image')]
    
    images=[]
    for path in paths:
        images.append(Image.open(path))
    image1_size = images[0].size
    
    new_image = Image.new('RGB',(4*image1_size[0], 4*image1_size[1]), (250,250,250))
    
    new_image.paste(images[0],(0,0))
    new_image.paste(images[1],(image1_size[0],0))
    new_image.paste(images[2],(2*image1_size[0],0))
    new_image.paste(images[3],(3*image1_size[0],0))
    new_image.paste(images[4],(0,image1_size[1]))
    new_image.paste(images[5],(image1_size[0],image1_size[1]))
    new_image.paste(images[6],(2*image1_size[0],image1_size[1]))
    new_image.paste(images[7],(3*image1_size[0],image1_size[1]))
    new_image.paste(images[8],(0,2*image1_size[1]))
    new_image.paste(images[9],(image1_size[0],2*image1_size[1]))
    new_image.paste(images[10],(2*image1_size[0],2*image1_size[1]))
    new_image.paste(images[11],(3*image1_size[0],2*image1_size[1]))
    new_image.paste(images[12],(0,3*image1_size[1]))
    new_image.paste(images[13],(image1_size[0],3*image1_size[1]))
    new_image.paste(images[14],(2*image1_size[0],3*image1_size[1]))
    new_image.paste(images[15],(3*image1_size[0],3*image1_size[1]))
    fname = "{}.jpg".format("{0:05d}".format(counted))
    savename = save_path +fname
    new_image.save(savename,"JPEG")
    #new_image.show()
    
```

```python
image_paths = [os.path.join('input', x) for x in os.listdir('input')]
counted=0
import math
import time

start = time.time()
math.factorial(100000)

for image_path in image_paths:
    if counted ==0:
        image_crop(image_path,'current_image/')
        next_image_paths = [os.path.join('current_image', x) for x in os.listdir('current_image')]
        for k in next_image_paths:
            low_res = cv2.imread(k, 1)
            low_res = cv2.cvtColor(low_res, cv2.COLOR_BGR2RGB)
            low_res = low_res / 255.0
            sr = model.predict(np.expand_dims(low_res, axis=0))[0]
            sr = (((sr + 1) / 2.) * 255).astype(np.uint8)
            sr = cv2.cvtColor(sr, cv2.COLOR_RGB2BGR)
            cv2.imwrite(os.path.join('transform_image', os.path.basename(k)), sr)
    else:
        image_crop(image_path,'input_image/')
        input_images = [os.path.join('input_image', x) for x in os.listdir('input_image')]
        current_images = [os.path.join('current_image', x) for x in os.listdir('current_image')]
        
        for i in range(16):
            result = compare_image(input_images[i],current_images[i])
            #print(input_images[i]+current_images[i])
            #print(result)
            if result == False:
                #current_image 변환
                img = cv2.imread(input_images[i])
                cv2.imwrite(current_images[i], img)
                #변환
                low_res = cv2.imread(input_images[i], 1)
                low_res = cv2.cvtColor(low_res, cv2.COLOR_BGR2RGB)
                low_res = low_res / 255.0
                sr = model.predict(np.expand_dims(low_res, axis=0))[0]
                sr = (((sr + 1) / 2.) * 255).astype(np.uint8)
                sr = cv2.cvtColor(sr, cv2.COLOR_RGB2BGR)
                cv2.imwrite(os.path.join('transform_image', os.path.basename(input_images[i])), sr)
                print("변환 완료"+"i")
    #transform image 폴더에 있는 분할된 이미지 합쳐서 mereged_image 폴더에 저장하는 함수
    concat(counted , 'merged_image/' )
    counted=counted+1
end = time.time()

print(f"{end - start:.5f} sec")
```
